```python
import numpy as np
import matplotlib.pyplot as plt
import cv2 as cv
```

```python
img = plt.imread("madrid.jpg")
plt.imshow(img)
```

```python
t, l, warna = img.shape
```

KONVERSI GAMBAR MENJADI GRAY TANPA MENGGUNAKAN LIBRARY

```python
hasil = np.zeros([t, l]).astype(int)
tmp = 0
for i in range(t):
    for j in range(l):
        for x in range(warna):
            tmp += img[i][j][x]

        hasil [i, j] = tmp/3
        tmp = 0
plt.imshow(hasil, cmap='gray')
```

KONVERSI GAMBAR MENJADI GRAY MENGGUNAKAN LIBRARY

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap = 'gray')
```

PERBANDINGAN KONVERSI GAMBAR MENJADI GRAY DENGAN LIBRARY DAN TANPA LIBRARY

```python
plt.figure(facecolor='white')
plt.subplot(121)
plt.imshow(gray_img, cmap='gray')
plt.title("Dengan Library")
plt.subplot(122)
plt.imshow(hasil, cmap='gray')
plt.title("Tanpa Library")
```
