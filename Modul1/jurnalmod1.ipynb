```python
import numpy as np #berfungsi untuk memanggil atau memasukkan library numpy yang diinisialisasikan dengan np
import matplotlib.pyplot as plt #berfungsi untuk memanggil atau memasukkan library matplotlib.pyplot yang diinisialisasikan dengan plt
import cv2 as cv #berfungsi untuk memanggil atau memasukkan library cv2 yang diinisialisasikan dengan cv
```

```python
img = plt.imread('madrid.jpg') #berfungsi untuk menampilkan gambar madrid.jpg
plt.imshow(img)
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
plt.imshow(gray_img, cmap = "gray") #berfungsi untuk merubah warna gambar menjadi gray atau abu-abu
```

```python
print(gray_img)
```

```python
t, l = gray_img.shape
print(t, l) #berfungsi untuk mengetahui jumlah piksel
```

```python
sumgray = np.zeros(256) #berfungsi untuk membuat array yang anyaknya 256
for  i in range(t):
    for j in range(l):
        val = img[i,j]
        sumgray[val] += 1 #perulangan ini untuk membaca semua perulangan yang akan dibaca dari baris pertama kolom pertama lalu akan menghitung jumlah dari angkanya di deret matriks tersebut lalu terus akan bertambah satu

print(sumgray) #berfungsi untuk mencetak angkanya
```

```python
grayvalue = [i for i in range(256)]
print(grayvalue)
```

```python
h = sumgray/t*l
plt.bar(grayvalue, h)
```

```python
kernel = np.array([[0, -1, 0],
                    [-1, 5, -1],
                    [0, -1, 0],
                    [1, 0, -1]]) #berisi nilai kernel yang diinput dalam bentuk array
konv = np.zeros(gray_img.shape) #menunjukkan bahwa padding yang digunakan 0
for i in range (1278):
    for j in range (1021):
        konv[i][j] = gray_img[i][j]*kernel[0][0] + gray_img[i][j+1]*kernel[0][1] + gray_img[i][j+2]*kernel[0][2] + gray_img[i+1][j]*kernel[1][0] + gray_img[i+1][j+1]*kernel[1][1] + gray_img[i+1][j+2]*kernel[1][2] + gray_img[i+2][j]*kernel[2][0] + gray_img[i+2][j+1]*kernel[2][1] + gray_img[i+1][j+2]*kernel[2][2] #jumlah dari perkalian dari penjumlahan ini akan disimpan dalam konv 
#konv disini yang akan membuat warna abu menjadi lebih jelas
print(konv)
```

```python
plt.imshow(konv, cmap = "gray") #berfungsi untuk menampilkan gambar
```

```python
pad = max(kernel.shape) #untuk membuat padding yang berisi nilai terbesar atau maksimum dari ukuran kernelnya
if pad%2==0: #jika padding modulus 2 atau bernilai genap maka paddingnya dibagi 2
    pad/=2
else: pad=(pad+1)/2 #jika tidak maka nilai paddingnya ditambah 1 dan dibagi 2
gray = np.pad(gray_img, int(pad)) #membuat matriks sesuai dengan ukuran padding, di mana ukuran padding nya diberi parameter gray_img dan nilai padding itu bertype data int
konv = np.zeros(gray_img.shape) #membuat matriks yang berisikan 0
konvClip = np.zeros(gray_img.shape) #membuat matriks yang berisikan 0
row, col = konv.shape #nilai kolom dan baris berisi nilai ukuran dari gambar yang sudah di konvolusi
rowKernel, colKernel = kernel.shape #nilai kolom dan baris berisi nilai ukuran dari kernel
for i in range (row):
    for j in range (col):
        for rk in range (rowKernel):
            for ck in range (colKernel):
                konv[i][j] += gray[i+rk][j+ck]*kernel[rk][ck] #jumlah dari perkalian dari penjumlahan ini akan disimpan dalam konv 
                                                                #konv disini yang akan membuat warna abu menjadi lebih jelas
                konvClip[i][j] = konv[i][j] #nilai dari konvolusi clipping berisi nilai dari konvolusi tanpa clipping
                if konvClip[i][j]<0: #jika nilai dari variabel konvClip yang dimana baris dan kolomnya terdapat kurang dari 0, maka akan dibulatkan menjadi 0
                    konvClip[i][j] = 0
                if konvClip[i][j]>255: #jika nilai dari variabel konvClip yang dimana baris dan kolomnya terdapat lebih dari 255, maka akan dibulatkan menjadi 255
                    konvClip[i][j] = 255

print(konvClip)
```

```python
plt.imshow(konvClip, cmap="gray") #berfungsi untuk menampilkan gambar konvolusi clipping
```
