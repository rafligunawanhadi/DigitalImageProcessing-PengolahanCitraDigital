```python
import numpy as np #berfungsi untuk memanggil import numpy yang diinisialkan dengan np
import matplotlib.pyplot as plt #untuk mempermudah secara umum matlotlib, pyplot disingkat menjadi plt
import matplotlib.image as mpimg #membaca gambar yang di import
import cv2 as cv
```

```python
img = mpimg.imread("incredible.jpg") #berfungsi untuk menampilkan gambar
plt.imshow(img)
```

```python
gray_img = cv.cvtColor(img, cv.COLOR_BGR2GRAY)
negatif = 255-gray_img
plt.imshow(gray_img, cmap="gray") #berfungsi untuk mengubah warna menjadi gray atau abu-abu
```

```python
#TRANSLASI
img = cv.imread("incredible.jpg", 0) #berfungsi untuk mengimport gambar, agar memberikan padding 0 pada gambar, sehingga gambar akan terlihat bergeser
print(img)
panjang = len(gray_img) #panjang sama dengan panjang dari gambar yang ditampilkan
img = np.array(gray_img) #berfungsi untuk melihat nilai matriks dari gambar yang akan ditampilkan
translasi=np.zeros((256,256)) #berfungsi untuk memberikan 1 array 0 pada matriks yang berukuran 256 x 256
for i in range (panjang-50):
    for j in range (panjang-50):
        translasi[i-40][j+30] = gray_img[i][j]
print(translasi) #berfungsi untuk menampilkan nilai translasi atau pergeseran indeks/letak
print(translasi)
plt.imshow(translasi, cmap="gray") #berfungsi untuk menampilkan gambar yang sudah mengalami pergeseran letak yang diubah menjadi grayscale
```

```python
#ROTASI
citra = cv.imread("incredible.jpg", 0)
x, y = citra.shape
```

```python
def rotasi(degree):
    rotasi = np.zeros((x, y)) #membuat matriks kosong
    angle = -degree*(np.pi/180) #merupakan rumus konversi derajat ke radian

    center_x = x // 2 #merupakan pusat gambar
    center_y = x // 2 #merupakan pusat gambar

    for i in range(x):
        for j in range(y):
            x2 = int((i - center_x)* np.cos(angle)-(j - center_y)* np.sin(angle)+ center_x)
            y2 = int((i - center_x)* np.sin(angle)+(j - center_y)* np.cos(angle)+ center_x)
            if 0 <= x2 < x and 0 <= y2 < y:
                rotasi[i][j] = citra[x2][y2]
    plt.imshow(rotasi, cmap="gray")

rotasi(170)
```

```python
#PENCERMINAN SUMBU X
cerminx = np.zeros(gray_img.shape) #berfungsi untuk memberikan nilai array kosong sebanyak panjang gambar yang ditampilkan
mir=panjang-1 #mir, mendeklarasikan bahwa setiap panjang akan dikurangi 1
for i in range(panjang): #perulangan pada baris pada nilai ukuran dari matriks
    for j in range(panjang): #perulangan pada kolom pada nilai ukuran dari matriks
        cerminx[i][j] = gray_img[mir][j] #merupakan rumus dari pencerminan sumbu x
    mir -= 1 #setelah melakukan perulangan, maka perulangan akan dihitung sudah melakukan perulangan dan akan melakukan perulangan yang lainnya
plt.imshow(cerminx, cmap="gray")
```

```python
#CERMIN SUMBU Y
cerminy = np.zeros(gray_img.shape) #berfungsi untuk memberikan nilai array kosong sebanyak panjang gambar yang ditampilkan
mir=panjang-1 #mir, mendeklarasikan bahwa setiap panjang akan dikurangi 1
for i in range(panjang): #perulangan pada baris pada nilai ukuran dari matriks
    for j in range(panjang): #perulangan pada kolom pada nilai ukuran dari matriks
        cerminy[j][i] = gray_img[j][mir] #merupakan rumus dari pencerminan sumbu y
    mir -= 1 #setelah melakukan perulangan, maka perulangan akan dihitung sudah melakukan perulangan dan akan melakukan perulangan yang lainnya
plt.imshow(cerminy, cmap="gray") #untuk menampilkan gambar
```

```python
#DILATASI
row, col = gray_img.shape
skala = 2
dilatasi = np.zeros((row*skala, col*skala))
for i in range(len(dilatasi)):
    for j in range(len(dilatasi)):
        dilatasi[i][j] = gray_img[int(i/2), int(j/2)]
print(dilatasi)
plt.imshow(dilatasi, cmap="gray")
```
